```python
import pandas as pd

crops_soil_types = {
    "Rice": "alluvial soils, black soils, deltaic soils",
    "Maize": "black soils, laterite soils, red soils",
    "Jute": "alluvial soils, deltaic soils",
    "Cotton": "black soils, red soils, laterite soils",
    "Coconut": "coastal sands, alluvial soils, black soils",
    "Papaya": "alluvial soils, black soils, red soils",
    "Orange": "laterite soils, red soils, black soils",
    "Apple": "mountain soils, forest soils",
    "Muskmelon": "alluvial soils, black soils",
    "Watermelon": "alluvial soils, black soils",
    "Grapes": "alluvial soils, black soils, red soils",
    "Mango": "laterite soils, red soils, black soils",
    "Banana": "alluvial soils, deltaic soils, coastal sands",
    "Pomegranate": "laterite soils, red soils, black soils",
    "Lentil": "black soils, alluvial soils",
    "Blackgram": "black soils, alluvial soils",
    "Mungbean": "black soils, alluvial soils",
    "Mothbeans": "black soils, alluvial soils",
    "Pigeonpeas": "black soils, alluvial soils",
    "Kidneybeans": "black soils, alluvial soils",
    "Chickpea": "black soils, alluvial soils",
    "Coffee": "mountain soils, forest soils"
}

# Get a list of all unique soil types
unique_soil_types = set()
for soil_types in crops_soil_types.values():
    soil_types = soil_types.split(", ")
    unique_soil_types |= set(soil_types)
unique_soil_types = list(unique_soil_types)

# Initialize a dataframe with one-hot encoded columns for each soil type
df1 = pd.DataFrame(columns=unique_soil_types)

# Populate the dataframe with one-hot encoded values for each crop
for crop, soil_types in crops_soil_types.items():
    soil_types = soil_types.split(", ")
    values = [1 if soil_type in soil_types else 0 for soil_type in unique_soil_types]
    df1 = df1.append(pd.Series(values, index=df1.columns, name=crop))

# Reset the index of the dataframe
df1 = df1.reset_index()
df1 = df1.rename(columns={"index": "Crop"})

print(df1)
```
